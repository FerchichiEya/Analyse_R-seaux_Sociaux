```python
import gzip
import networkx as nx

def demon_algorithm(G):
    communities = []
    visited = set()

    for node in G.nodes():
        if node not in visited:
            community = set()
            neighbors = set(G.neighbors(node))

            while neighbors:
                n = neighbors.pop()
                community.add(n)
                visited.add(n)
                neighbors |= set(G.neighbors(n)) - visited

            communities.append(community)

    return communities

with gzip.open("ca-AstroPh.txt.gz", "rt") as f:
    data = f.readlines()

G = nx.Graph()

for line in data:
    if line.startswith("#"):
        continue
    edge = line.strip().split()
    author1 = edge[0]
    author2 = edge[1]
    G.add_edge(author1, author2)

demon_communities = demon_algorithm(G)

print("\nNombre de communautés selon Demon:", len(demon_communities))
print("Communautés selon Demon:")
for i, community in enumerate(demon_communities):
    print(f"Communauté {i+1}: {community}")
```
